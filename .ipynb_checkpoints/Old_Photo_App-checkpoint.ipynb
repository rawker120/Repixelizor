```python
import flet as ft

def main(page: ft.Page):
    page.title = "Old Image to Life"
    page.theme_mode = ft.ThemeMode.LIGHT
    page.padding = 20
    page.update()
    page.scroll=True
    path="https://replicate.delivery/pbxt/cYi7SwsnRraIHx0tAdtEnw2lgQqfNn8fqbcmBVdu9HD7zkjQA/out.png"
    path_update=""
    
    
    
    def pick_files_result(e: ft.FilePickerResultEvent):
        selected_files.value = (
            ", ".join(map(lambda f: f.name, e.files)) if e.files else "Cancelled!"
        )
        selected_files.update()
        file_path.value = e.files[0].path
        file_path.update()
#         page.remove(img, images)
        update(e.files[0].path)
        path=e.files[0].path
#         path.update()
        
    def update(path):
        img.src=path
        img.update()
        process_button.disabled=False
        process_button.update()
        
        
    def output(path):
        upload_button.disabled=True
        upload_button.update()
        process_button.disabled=True
        process_button.update()
        pb.visible=True
        pb.update()
        progress_txt.visible=True
        progress_txt.update()
        inputs = {
    # input image.
        'image': open(path, "rb"),

        # whether the input image is high-resolution
        'HR': True,

        # whether the input image is scratched
        'with_scratch': True,}

# https://replicate.com/microsoft/bringing-old-photos-back-to-life/versions/c75db81db6cbd809d93cc3b7e7a088a351a3349c9fa02b6d393e35e0d51ba799#output-schema
        output = version.predict(**inputs)
        img2.src=output
        img2.update()
        pb.visible=False
        pb.update()
        progress_txt.visible=False
        progress_txt.update()
        upload_button.disabled=False
        upload_button.update()
        process_button.disabled=False
        process_button.update()

    pick_files_dialog = ft.FilePicker(on_result=pick_files_result)
    selected_files = ft.Text()
    file_path = ft.Text(value="Selected file path", expand=1)
    page.overlay.append(pick_files_dialog)
    upload_button=ft.ElevatedButton(
                    "Pick files",
                    icon=ft.icons.UPLOAD_FILE,
                    on_click=lambda _: pick_files_dialog.pick_files(
                        allow_multiple=True
                    ),
                )
    process_button=ft.ElevatedButton(
                    "Process",
                    icon=ft.icons.API,
                    on_click=lambda _: output(file_path.value),
                    disabled=True
                )
    page.add(
        ft.Row(
            [upload_button
                ,
                selected_files,file_path,process_button
            ]
        )
    )
    img = ft.Image(
    src=path,
    width=400,
    height=800,
    fit=ft.ImageFit.CONTAIN,)
    img2 = ft.Image(
    src=path_update,
    width=400,
    height=800,
    fit=ft.ImageFit.CONTAIN,)
    pb = ft.ProgressBar(width=400, color="amber", bgcolor="#eeeeee",visible=False)
    progress_txt=ft.Text("Model Running", style="headlineSmall",visible=False)
    images = ft.Column(expand=1, wrap=False, scroll="always")
    page.add(ft.Row([progress_txt,pb]),ft.Row([img,img2]))
    

    page.update()

ft.app(target=main)
```
